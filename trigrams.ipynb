```python
# Selecting random items from lists.
import random

# Efficient data structures.
import collections
```

Open File

```python
#These are the characters we will be using.
chars = 'abcdefghijklmnopqrstuvwxyz .'
```

```python
with open('dracula.txt', 'r', encoding='utf-8') as file:
  # Read the whole file into a string.
  dracula = file.read()
```

Preprocess File Contents

```python
def create_trigram_model(text):
    trigram_counts = {}  # Dictionary to store trigram counts
    
    # Loop through the text and extract trigrams
    for i in range(len(text) - 2):
        trigram = text[i:i+3]  # Extract a trigram of three characters
        if trigram in trigram_counts:
            trigram_counts[trigram] += 1  # Increment count if trigram exists
        else:
            trigram_counts[trigram] = 1  # Initialize count if trigram is new
    
    return trigram_counts
```

```python
#Change everything to lower case.
dracula = dracula.lower()
```

```python
#remove unwanted character.
cleaned = ''.join(c for c in dracula if c in chars)
```

```python
# Count the frequency of each character.
counts = collections.Counter(cleaned)
```

```python
# Show, sorted.
counts.most_common()
```

```python
# Create the trigram model
trigram_model = create_trigram_model(cleaned)
```

```python
# Example output of the trigram model
for trigram, count in trigram_model.items():
    print(f"'{trigram}': {count}")
```
