Preprocess File Contents

```python
def create_trigram_model(text, trigram_counts):
    # Loop through the text and extract trigrams
    for i in range(len(text) - 2):
        trigram = text[i:i+3]  # Extract a trigram of three characters
        if trigram in trigram_counts:
            trigram_counts[trigram] += 1  # Increment count if trigram exists
        else:
            trigram_counts[trigram] = 1  # Initialize count if trigram is new
    
    return trigram_counts  # Return the updated dictionary
```

```python
def remove_preamble_and_postamble(text):
    start_marker = "*** START OF THIS PROJECT GUTENBERG EBOOK"
    end_marker = "*** END OF THIS PROJECT GUTENBERG EBOOK"
    start_idx = text.find(start_marker)
    end_idx = text.find(end_marker)

    if start_idx != -1:
        text = text[start_idx + len(start_marker):]
    if end_idx != -1:
        text = text[:end_idx]

    return text
```

```python
#remove unwanted character.
def clean_text(text):
    chars = 'abcdefghijklmnopqrstuvwxyz .'  # Allowed characters
    cleaned = ''.join(c for c in text.lower() if c in chars)  # Convert to lowercase and filter out unwanted characters
    return cleaned.upper()  # Return the cleaned text in uppercase
```

```python
#Lists of books to process
books = ['dracula.txt', 'metamorphosis.txt','alice.txt', 'thecallofcthulhu.txt', 'theiliad.txt']
```

```python
# Initialize an empty dictionary to store the combined trigram counts
combined_trigram_counts = {}
```

Processing Texts to Build the Combined Trigram Model

```python
# Process each book
for book in books:
    try:
        with open(book, 'r', encoding='utf-8') as file:
            text = file.read()
            text = remove_preamble_and_postamble(text)  # Trim preamble and postamble
            cleaned_text = clean_text(text)  # Clean and convert to uppercase
            print(f"Successfully read '{book}' (first 500 characters):\n{text[:500]}\n")
    except Exception as e:
        print(f"Error reading '{book}': {e}")
        continue  # Skip this book if there's an error
    
    
    # Update the trigram model with data from the current book
    try:
        combined_trigram_counts = create_trigram_model(cleaned_text, combined_trigram_counts)
        print(f"Trigrams updated for '{book}'\n")
    except Exception as e:
        print(f"Error updating trigram model for '{book}': {e}")
        continue  # Skip this book if there's an error
```

Sorting Trigram Counts in Descending Order

```python
# Create a list of (trigram, count) from the dictionary
trigram_list = list(combined_trigram_counts.items())  # This turns the dictionary into a list of tuples

# Sort the list of tuples based on the count (the second element of each tuple)
# We'll use a loop to explain the sorting logic
for i in range(len(trigram_list)):
    for j in range(i + 1, len(trigram_list)):
        # Compare the counts of two trigrams, and swap them if they are out of order
        if trigram_list[i][1] < trigram_list[j][1]:
            trigram_list[i], trigram_list[j] = trigram_list[j], trigram_list[i]

# Print the sorted trigrams
for trigram, count in trigram_list[:500]:
    print(f"'{trigram}': {count}")
```

Task 2

This section groups trigrams from the combined trigram model (combined_trigram_counts) by their first two characters, forming a dictionary called trigram_dict.

```python
# Initialize an empty dictionary to store the trigrams grouped by their first two characters.
trigram_dict = {}

# Loop through each trigram and its count in the combined trigram counts dictionary.
for trigram, count in combined_trigram_counts.items():
    # Extract the first two characters of the trigram to use as a prefix.
    prefix = trigram[:2]
    
    # If the prefix doesn't already exist in the trigram_dict, initialize it with an empty list.
    if prefix not in trigram_dict:
        trigram_dict[prefix] = []
    
    # Append the continuation (the third character of the trigram) and its count to the list for this prefix.
    trigram_dict[prefix].append((trigram[2], count))
```

```python
for prefix, continuations in trigram_dict.items():
    print(f"Prefix '{prefix}' has continuations: {continuations}")
```

This function, weighted_choice, selects an item randomly from a list of (item, weight) tuples, where each item's probability of being chosen is proportional to its weight.

```python
def weighted_choice(choices):
    # choices is a list of (item, weight) tuples
    total = sum(weight for item, weight in choices)
    r = random.uniform(0, total)
    upto = 0
    for item, weight in choices:
        if upto + weight > r:
            return item
        upto += weight
```

```python
generated_text = "TH"

for _ in range(10000 - len(generated_text)):
    # Get the last two characters
    prefix = generated_text[-2:]
    
    # Check if the prefix exists in the trigram dictionary
    if prefix in trigram_dict:
        next_char = weighted_choice(trigram_dict[prefix])
        generated_text += next_char
    else:
        # If no trigrams start with this prefix, break the loop
        break
```

```python
# Convert the final generated text to uppercase for display or saving
with open('generated_text.txt', 'w') as file:
    file.write(generated_text.upper())

print(f"Generated text (first 500 characters):\n{generated_text[:500].upper()}")
```
